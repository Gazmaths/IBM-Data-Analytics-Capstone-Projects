### Step 1: Import Libraries and Load Data


- Import the `pandas`, `matplotlib.pyplot`, and `seaborn` libraries.


- You will begin with loading the dataset. You can use the pyfetch method if working on JupyterLite. Otherwise, you can use pandas' read_csv() function directly on their local machines or cloud environments.


```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the Stack Overflow survey dataset
data_url = 'https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv'
df = pd.read_csv(data_url)

# Display the first few rows of the dataset
df.head()
```

### Step 2: Examine the Structure of the Data


- Display the column names, data types, and summary information to understand the data structure.

- Objective: Gain insights into the dataset's shape and available variables.


```python
## Write your code here
# 1. Display the first few rows
print("Preview of data:")
print(df.head())

# 2. Check the number of rows and columns
print("\nShape of the dataset:")
print(df.shape)  # (rows, columns)

# 3. List all column names
print("\nColumn names:")
print(df.columns.tolist())

# 4. Show data types and non-null counts
print("\nData types and non-null counts:")
print(df.info())

# 5. Display summary statistics for numerical columns
print("\nSummary statistics:")
print(df.describe())
```

### Step 3: Handle Missing Data


- Identify missing values in the dataset.

- Impute or remove missing values as necessary to ensure data completeness.



```python
## Write your code here
# 1. Count of missing values in each column
missing_counts = df.isnull().sum()
print("Missing values per column:")
print(missing_counts)

# 2. Percentage of missing values
missing_percentage = (df.isnull().mean() * 100).round(2)
print("\nPercentage of missing values:")
print(missing_percentage)
```

### Step 4: Analyze Key Columns


- Examine key columns such as `Employment`, `JobSat` (Job Satisfaction), and `YearsCodePro` (Professional Coding Experience).

- **Instruction**: Calculate the value counts for each column to understand the distribution of responses.



```python
## Write your code here
# Employment distribution
print("=== Employment Type Distribution ===")
print(df['Employment'].value_counts(dropna=False))
print()

# Job Satisfaction distribution
print("=== Job Satisfaction Distribution ===")
print(df['JobSat'].value_counts(dropna=False))
print()

# Years of Professional Coding Experience distribution
print("=== YearsCodePro Distribution ===")
print(df['YearsCodePro'].value_counts(dropna=False))
print()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for Employment
df['Employment'].value_counts().plot(kind='bar', title='Employment Type Distribution', figsize=(10,4))
plt.ylabel('Count')
plt.xlabel('Employment Type')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar plot for Job Satisfaction
df['JobSat'].value_counts().plot(kind='bar', title='Job Satisfaction Distribution', figsize=(10,4), color='teal')
plt.ylabel('Count')
plt.xlabel('Job Satisfaction')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

# Bar plot for YearsCodePro
df['YearsCodePro'].value_counts().sort_index().plot(kind='bar', title='Professional Coding Experience', figsize=(10,4), color='orange')
plt.ylabel('Count')
plt.xlabel('Years of Experience')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 5: Visualize Job Satisfaction (Focus on JobSat)


- Create a pie chart or KDE plot to visualize the distribution of `JobSat`.

- Provide an interpretation of the plot, highlighting key trends in job satisfaction.


```python
## Write your code here
# Drop missing values in JobSat
df_clean = df.dropna(subset=['JobSat'])
import matplotlib.pyplot as plt

# Value counts for JobSat
job_sat_counts = df_clean['JobSat'].value_counts()

# Pie chart
plt.figure(figsize=(8, 8))
plt.pie(job_sat_counts, labels=job_sat_counts.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Job Satisfaction Distribution')
plt.axis('equal')
plt.show()

```

```python
import seaborn as sns

# Drop missing values in JobSatPoints_1
df_kde = df.dropna(subset=['JobSatPoints_1'])

# Plot KDE
plt.figure(figsize=(8, 5))
sns.kdeplot(df_kde['JobSatPoints_1'], fill=True, color='skyblue')
plt.title('KDE Plot of Job Satisfaction')
plt.xlabel('Job Satisfaction Score')
plt.ylabel('Density')
plt.grid(True)
plt.show()
```

### Step 6: Programming Languages Analysis


- Compare the frequency of programming languages in `LanguageHaveWorkedWith` and `LanguageWantToWorkWith`.
  
- Visualize the overlap or differences using a Venn diagram or a grouped bar chart.


```python
## Write your code here
# Drop rows with missing values in the relevant columns
df_lang = df.dropna(subset=['LanguageHaveWorkedWith', 'LanguageWantToWorkWith'])

# Split each semicolon-separated string into lists
have_worked = df_lang['LanguageHaveWorkedWith'].str.split(';').explode().str.strip()
want_to_work = df_lang['LanguageWantToWorkWith'].str.split(';').explode().str.strip()

# Count frequencies
have_counts = have_worked.value_counts()
want_counts = want_to_work.value_counts()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Combine into a single DataFrame
lang_df = pd.DataFrame({
    'HaveWorkedWith': have_counts,
    'WantToWorkWith': want_counts
}).fillna(0)

# Select top languages (adjust as needed)
top_langs = lang_df.sum(axis=1).sort_values(ascending=False).head(10)
lang_df_top = lang_df.loc[top_langs.index]

# Plot
lang_df_top.plot(kind='bar', figsize=(12,6))
plt.title('Comparison of Languages Worked With vs. Wanted to Work With')
plt.ylabel('Count')
plt.xlabel('Programming Language')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

### Step 7: Analyze Remote Work Trends


- Visualize the distribution of RemoteWork by region using a grouped bar chart or heatmap.


```python
## Write your code here
import pandas as pd

# Drop rows with missing values in either column
df_remote = df.dropna(subset=['RemoteWork', 'Country'])  # or 'Region' if available

# Create a cross-tabulation of counts
remote_region_counts = pd.crosstab(df_remote['Country'], df_remote['RemoteWork'])

# (Optional) Normalize by row to get percentages
# remote_region_counts = remote_region_counts.div(remote_region_counts.sum(axis=1), axis=0)
```

```python
import matplotlib.pyplot as plt

# Select top N countries by total respondents
top_countries = remote_region_counts.sum(axis=1).sort_values(ascending=False).head(10).index
remote_subset = remote_region_counts.loc[top_countries]

# Plot
remote_subset.plot(kind='bar', figsize=(12, 6))
plt.title('Distribution of Remote Work by Country')
plt.ylabel('Number of Respondents')
plt.xlabel('Country')
plt.xticks(rotation=45)
plt.legend(title='Remote Work Type')
plt.tight_layout()
plt.show()
```

```python
import seaborn as sns

plt.figure(figsize=(12, 8))
sns.heatmap(remote_region_counts, cmap='Blues', linewidths=0.5, annot=True, fmt='d')
plt.title('Remote Work Distribution by Country')
plt.xlabel('Remote Work Type')
plt.ylabel('Country')
plt.tight_layout()
plt.show()
```

### Step 8: Correlation between Job Satisfaction and Experience


- Analyze the correlation between overall job satisfaction (`JobSat`) and `YearsCodePro`.
  
- Calculate the Pearson or Spearman correlation coefficient.


```python
import numpy as np
def convert_experience(val):
    if pd.isna(val):
        return np.nan
    elif val == 'Less than 1 year':
        return 0.5
    elif val == 'More than 50 years':
        return 51
    else:
        try:
            return float(val)
        except:
            return np.nan
```

```python
df['YearsCodeProNum'] = df['YearsCodePro'].apply(convert_experience)
df_corr = df[['JobSat', 'YearsCodeProNum']].dropna()
pearson_corr = df_corr.corr(method='pearson')
print("📈 Pearson Correlation:\n", pearson_corr)

spearman_corr = df_corr.corr(method='spearman')
print("📈 Spearman Correlation:\n", spearman_corr)


```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.scatterplot(data=df_corr, x='YearsCodeProNum', y='JobSat', alpha=0.5)
sns.regplot(data=df_corr, x='YearsCodeProNum', y='JobSat', scatter=False, color='red')
plt.title('Job Satisfaction vs. Years of Professional Coding Experience')
plt.xlabel('Years of Professional Coding Experience')
plt.ylabel('Job Satisfaction Score')
plt.show()
```

### Step 9: Cross-tabulation Analysis (Employment vs. Education Level)


- Analyze the relationship between employment status (`Employment`) and education level (`EdLevel`).

- **Instruction**: Create a cross-tabulation using `pd.crosstab()` and visualize it with a stacked bar plot if possible.


```python
## Write your code here
import pandas as pd

# Create cross-tabulation (counts of Employment status per Education level)
crosstab = pd.crosstab(df['EdLevel'], df['Employment'])

# (Optional) To see proportions within each education level, normalize by row:
# crosstab = pd.crosstab(df['EdLevel'], df['Employment'], normalize='index')

print(crosstab)

import matplotlib.pyplot as plt

# Plot stacked bar chart
crosstab.plot(kind='bar', stacked=True, figsize=(12,6), colormap='tab20')

plt.title('Employment Status Distribution by Education Level')
plt.xlabel('Education Level')
plt.ylabel('Number of Respondents')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Employment Status', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
