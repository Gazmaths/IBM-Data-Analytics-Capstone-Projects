### Task 1: Exploring Data Distributions Using Bubble Plots


#### 1. Bubble Plot for Age vs. Frequency of Participation


- Visualize the relationship between respondents’ age and their participation frequency (`SOPartFreq`) using a bubble plot.

- Use the size of the bubbles to represent their job satisfaction (`JobSat`).


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the data
df = pd.read_csv("survey-data.csv")

# Drop rows with missing values in the required columns
bubble_df = df[['Age', 'SOPartFreq', 'JobSat']].dropna()

# Optional: Convert categorical 'Age' and 'SOPartFreq' to numeric codes for plotting
bubble_df['AgeCode'] = bubble_df['Age'].astype('category').cat.codes
bubble_df['SOPartFreqCode'] = bubble_df['SOPartFreq'].astype('category').cat.codes

# Normalize JobSat for bubble size scaling (optional for better visual effect)
bubble_df['JobSatSize'] = bubble_df['JobSat'] / bubble_df['JobSat'].max() * 1000

# Plotting
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    x=bubble_df['AgeCode'],
    y=bubble_df['SOPartFreqCode'],
    s=bubble_df['JobSatSize'],  # Bubble size
    alpha=0.6,
    c=bubble_df['JobSat'],
    cmap='viridis'
)

# Axis labels
plt.xlabel('Age')
plt.ylabel('Participation Frequency (SOPartFreq)')
plt.title('Age vs SOPartFreq with Job Satisfaction as Bubble Size')

# Create legend for JobSat
cbar = plt.colorbar(scatter)
cbar.set_label('Job Satisfaction')

# Replace ticks with original category labels
plt.xticks(ticks=range(len(bubble_df['Age'].astype('category').cat.categories)),
           labels=bubble_df['Age'].astype('category').cat.categories,
           rotation=45)

plt.yticks(ticks=range(len(bubble_df['SOPartFreq'].astype('category').cat.categories)),
           labels=bubble_df['SOPartFreq'].astype('category').cat.categories)

plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Compensation vs. Job Satisfaction


-Visualize the relationship between yearly compensation (`ConvertedCompYearly`) and job satisfaction (`JobSat`).

- Use the size of the bubbles to represent respondents’ age.


```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Select necessary columns and drop missing values
bubble_df = df[['ConvertedCompYearly', 'JobSat', 'Age']].dropna()

# Remove outliers from ConvertedCompYearly using IQR
Q1 = bubble_df['ConvertedCompYearly'].quantile(0.25)
Q3 = bubble_df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define upper and lower bounds
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter the data
bubble_df = bubble_df[(bubble_df['ConvertedCompYearly'] >= lower_bound) & 
                      (bubble_df['ConvertedCompYearly'] <= upper_bound)]

# Convert categorical Age to numeric codes
bubble_df['AgeCode'] = bubble_df['Age'].astype('category').cat.codes

# Normalize Age for bubble size
bubble_df['AgeSize'] = bubble_df['AgeCode'] / bubble_df['AgeCode'].max() * 1000

# Plotting
plt.figure(figsize=(12, 8))
scatter = plt.scatter(
    x=bubble_df['ConvertedCompYearly'],
    y=bubble_df['JobSat'],
    s=bubble_df['AgeSize'],
    alpha=0.5,
    c=bubble_df['AgeCode'],
    cmap='coolwarm'
)

plt.xlabel('Yearly Compensation (USD)')
plt.ylabel('Job Satisfaction')
plt.title('Compensation vs Job Satisfaction (Outliers Removed, Bubble Size = Age Group)')

cbar = plt.colorbar(scatter)
cbar.set_label('Age Group (Encoded)')

plt.grid(True)
plt.tight_layout()
plt.show()

```

### Task 2: Analyzing Relationships Using Bubble Plots


#### 1. Bubble Plot of Technology Preferences by Age

- Visualize the popularity of programming languages respondents have worked with (`LanguageHaveWorkedWith`) across age groups.

- Use bubble size to represent the frequency of each language.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Drop rows with missing values in Age or LanguageHaveWorkedWith
df_lang = df[['Age', 'LanguageHaveWorkedWith']].dropna()

# Explode the semicolon-separated languages
df_lang['LanguageHaveWorkedWith'] = df_lang['LanguageHaveWorkedWith'].str.split(';')
df_exploded = df_lang.explode('LanguageHaveWorkedWith')

# Count frequencies for each Age-Language combination
lang_counts = df_exploded.groupby(['Age', 'LanguageHaveWorkedWith']).size().reset_index(name='Count')

# Convert Age and Language to category codes for plotting
lang_counts['AgeCode'] = lang_counts['Age'].astype('category').cat.codes
lang_counts['LangCode'] = lang_counts['LanguageHaveWorkedWith'].astype('category').cat.codes

# Normalize Count for bubble size
lang_counts['Size'] = lang_counts['Count'] / lang_counts['Count'].max() * 1000

# Plotting
plt.figure(figsize=(14, 10))
scatter = plt.scatter(
    x=lang_counts['AgeCode'],
    y=lang_counts['LangCode'],
    s=lang_counts['Size'],
    alpha=0.6,
    c=lang_counts['Count'],
    cmap='plasma'
)

# Customize axis ticks to show actual labels
age_labels = lang_counts['Age'].astype('category').cat.categories
lang_labels = lang_counts['LanguageHaveWorkedWith'].astype('category').cat.categories

plt.xticks(ticks=range(len(age_labels)), labels=age_labels, rotation=45)
plt.yticks(ticks=range(len(lang_labels)), labels=lang_labels)

# Add labels and title
plt.xlabel('Age Group')
plt.ylabel('Programming Language')
plt.title('Popularity of Programming Languages by Age (Bubble Size = Frequency)')

# Colorbar for frequency
cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.tight_layout()
plt.grid(True) 

plt.show()
```

```python
#### 2. Bubble Plot for Preferred Databases vs. Job Satisfaction

- Explore the relationship between preferred databases (`DatabaseWantToWorkWith`) and job satisfaction.

- Use bubble size to indicate the number of respondents for each database.
```

```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Drop rows with missing values in the relevant columns
df_db = df[['DatabaseWantToWorkWith', 'JobSat']].dropna()

# Split the semicolon-separated databases
df_db['DatabaseWantToWorkWith'] = df_db['DatabaseWantToWorkWith'].str.split(';')
df_exploded = df_db.explode('DatabaseWantToWorkWith')

# Group by database and job satisfaction level, then count
db_counts = df_exploded.groupby(['DatabaseWantToWorkWith', 'JobSat']).size().reset_index(name='Count')

# Convert categories to numerical codes for plotting
db_counts['DBCode'] = db_counts['DatabaseWantToWorkWith'].astype('category').cat.codes
db_counts['JobSatCode'] = db_counts['JobSat'].astype('category').cat.codes

# Normalize count for bubble size
db_counts['Size'] = db_counts['Count'] / db_counts['Count'].max() * 1000

# Plotting
plt.figure(figsize=(14, 10))
scatter = plt.scatter(
    x=db_counts['JobSatCode'],
    y=db_counts['DBCode'],
    s=db_counts['Size'],
    alpha=0.6,
    c=db_counts['Count'],
    cmap='viridis'
)

# Set ticks to actual category names
job_sat_labels = db_counts['JobSat'].astype('category').cat.categories
db_labels = db_counts['DatabaseWantToWorkWith'].astype('category').cat.categories

plt.xticks(ticks=range(len(job_sat_labels)), labels=job_sat_labels, rotation=45)
plt.yticks(ticks=range(len(db_labels)), labels=db_labels)

# Add labels and colorbar
plt.xlabel('Job Satisfaction')
plt.ylabel('Preferred Database')
plt.title('Preferred Databases vs. Job Satisfaction (Bubble Size = Respondent Count)')

cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.tight_layout()
plt.grid(True)
plt.show()


#### Task 3: Comparing Data Using Bubble Plots
```

#### 1. Bubble Plot for Compensation Across Developer Roles

- Visualize compensation (`ConvertedCompYearly`) across different developer roles (`DevType`).

- Use bubble size to represent job satisfaction.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Drop rows with missing values
df_roles = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()
df_roles.head()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the dataset
df = pd.read_csv("survey-data.csv")

# Select relevant columns and drop rows with missing values
df_roles = df[['DevType', 'ConvertedCompYearly', 'JobSat']].dropna()

# Explode DevType because respondents can have multiple roles separated by semicolons
df_roles['DevType'] = df_roles['DevType'].str.split(';')
df_exploded = df_roles.explode('DevType')

# Remove compensation outliers using the IQR method
Q1 = df_exploded['ConvertedCompYearly'].quantile(0.25)
Q3 = df_exploded['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

df_clean = df_exploded[
    (df_exploded['ConvertedCompYearly'] >= lower_bound) &
    (df_exploded['ConvertedCompYearly'] <= upper_bound)
]

# Group by developer role and calculate average compensation and average job satisfaction
agg_df = df_clean.groupby('DevType').agg({
    'ConvertedCompYearly': 'mean',
    'JobSat': 'mean',
    'DevType': 'count'  # Count respondents per role
}).rename(columns={'DevType': 'Count'}).reset_index()

# Normalize JobSat for bubble size (scale max to 1000)
agg_df['Size'] = agg_df['JobSat'] / agg_df['JobSat'].max() * 1000

# Sort roles by average compensation (descending) for better visualization
agg_df = agg_df.sort_values('ConvertedCompYearly', ascending=False)

# Plot
plt.figure(figsize=(14, 10))
scatter = plt.scatter(
    x=agg_df['DevType'],
    y=agg_df['ConvertedCompYearly'],
    s=agg_df['Size'],
    alpha=0.6,
    c=agg_df['JobSat'],
    cmap='coolwarm',
    edgecolor='k',
    linewidth=0.5
)

plt.xticks(rotation=90)
plt.xlabel('Developer Role')
plt.ylabel('Average Compensation (USD)')
plt.title('Average Compensation by Developer Role\nBubble Size and Color = Avg Job Satisfaction')

cbar = plt.colorbar(scatter)
cbar.set_label('Average Job Satisfaction')

plt.tight_layout()
plt.grid(True)
plt.show()
```

#### 2. Bubble Plot for Collaboration Tools by Age

- Visualize the relationship between the collaboration tools used (`NEWCollabToolsHaveWorkedWith`) and age groups.

- Use bubble size to represent the frequency of tool usage.


```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load data
df = pd.read_csv("survey-data.csv")

# Select relevant columns and drop missing values
df_tools = df[['Age', 'NEWCollabToolsHaveWorkedWith']].dropna()

# Explode the semicolon-separated collaboration tools
df_tools['NEWCollabToolsHaveWorkedWith'] = df_tools['NEWCollabToolsHaveWorkedWith'].str.split(';')
df_exploded = df_tools.explode('NEWCollabToolsHaveWorkedWith')

# Count frequency by Age and Tool
freq_df = df_exploded.groupby(['Age', 'NEWCollabToolsHaveWorkedWith']).size().reset_index(name='Count')

# Convert Age and Tool to categorical codes for plotting
freq_df['AgeCode'] = freq_df['Age'].astype('category').cat.codes
freq_df['ToolCode'] = freq_df['NEWCollabToolsHaveWorkedWith'].astype('category').cat.codes

# Normalize count for bubble size
freq_df['Size'] = freq_df['Count'] / freq_df['Count'].max() * 1000

# Plot
plt.figure(figsize=(14,10))
scatter = plt.scatter(
    x=freq_df['AgeCode'],
    y=freq_df['ToolCode'],
    s=freq_df['Size'],
    alpha=0.6,
    c=freq_df['Count'],
    cmap='plasma'
)

# Set x and y ticks to original labels
age_labels = freq_df['Age'].astype('category').cat.categories
tool_labels = freq_df['NEWCollabToolsHaveWorkedWith'].astype('category').cat.categories

plt.xticks(ticks=range(len(age_labels)), labels=age_labels, rotation=45)
plt.yticks(ticks=range(len(tool_labels)), labels=tool_labels)

plt.xlabel('Age Group')
plt.ylabel('Collaboration Tool')
plt.title('Collaboration Tools Used by Age Group (Bubble Size = Frequency)')

cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.grid(True)
plt.tight_layout()
plt.show()
```

### Task 4: Visualizing Technology Trends Using Bubble Plots


#### 1. Bubble Plot for Preferred Web Frameworks vs. Job Satisfaction

- Explore the relationship between preferred web frameworks (`WebframeWantToWorkWith`) and job satisfaction.

- Use bubble size to represent the number of respondents.



```python
##Write your code here
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Select relevant columns and drop missing data
df_web = df[['WebframeWantToWorkWith', 'JobSat']].dropna()

# Explode the semicolon-separated list of web frameworks
df_web['WebframeWantToWorkWith'] = df_web['WebframeWantToWorkWith'].str.split(';')
df_exploded = df_web.explode('WebframeWantToWorkWith')

# Group by framework and job satisfaction, count respondents
counts = df_exploded.groupby(['WebframeWantToWorkWith', 'JobSat']).size().reset_index(name='Count')

# Convert categories to numeric codes for plotting
counts['JobSatCode'] = counts['JobSat'].astype('category').cat.codes
counts['FrameworkCode'] = counts['WebframeWantToWorkWith'].astype('category').cat.codes

# Normalize count for bubble size
counts['Size'] = counts['Count'] / counts['Count'].max() * 1000

# Plot
plt.figure(figsize=(14, 10))
scatter = plt.scatter(
    x=counts['JobSatCode'],
    y=counts['FrameworkCode'],
    s=counts['Size'],
    alpha=0.6,
    c=counts['Count'],
    cmap='viridis'
)

# Map ticks back to original labels
job_sat_labels = counts['JobSat'].astype('category').cat.categories
framework_labels = counts['WebframeWantToWorkWith'].astype('category').cat.categories

plt.xticks(ticks=range(len(job_sat_labels)), labels=job_sat_labels, rotation=45)
plt.yticks(ticks=range(len(framework_labels)), labels=framework_labels)

plt.xlabel('Job Satisfaction')
plt.ylabel('Preferred Web Framework')
plt.title('Preferred Web Frameworks vs Job Satisfaction (Bubble Size = Number of Respondents)')

cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.grid(True)
plt.tight_layout()
plt.show()
```

#### 2. Bubble Plot for Admired Technologies Across Countries

- Visualize the distribution of admired technologies (`LanguageAdmired`) across different countries (`Country`).

- Use bubble size to represent the frequency of admiration.



```python
import pandas as pd
import matplotlib.pyplot as plt

# Load dataset
df = pd.read_csv("survey-data.csv")

# Select relevant columns and drop missing values
df_admired = df[['Country', 'LanguageAdmired']].dropna()

# Split the semicolon-separated string into lists
df_admired['LanguageAdmired'] = df_admired['LanguageAdmired'].str.split(';')

# Explode the list to separate rows
df_exploded = df_admired.explode('LanguageAdmired')

# Strip whitespace from tech names (optional but recommended)
df_exploded['LanguageAdmired'] = df_exploded['LanguageAdmired'].str.strip()

# Group by Country and Technology, count frequency
freq_df = df_exploded.groupby(['Country', 'LanguageAdmired']).size().reset_index(name='Count')

# Convert categories to numeric codes for plotting
freq_df['CountryCode'] = freq_df['Country'].astype('category').cat.codes
freq_df['LangCode'] = freq_df['LanguageAdmired'].astype('category').cat.codes

# Normalize bubble sizes
freq_df['Size'] = freq_df['Count'] / freq_df['Count'].max() * 1000

# Plot
plt.figure(figsize=(20, 15))
scatter = plt.scatter(
    x=freq_df['CountryCode'],
    y=freq_df['LangCode'],
    s=freq_df['Size'],
    alpha=0.6,
    c=freq_df['Count'],
    cmap='plasma'
)

# Label axes with original category names
country_labels = freq_df['Country'].astype('category').cat.categories
lang_labels = freq_df['LanguageAdmired'].astype('category').cat.categories

plt.xticks(ticks=range(len(country_labels)), labels=country_labels, rotation=90)
plt.yticks(ticks=range(len(lang_labels)), labels=lang_labels)

plt.xlabel('Country')
plt.ylabel('Admired Technology')
plt.title('Admired Technologies Across Countries (Bubble Size = Frequency)')

cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.grid(True)
plt.tight_layout()
plt.show()

```

```python
# Calculate total respondents per country
country_totals = freq_df.groupby('Country')['Count'].sum().reset_index()

# Pick top 20 countries by respondent count
top_countries = country_totals.nlargest(20, 'Count')['Country']

# Filter freq_df to only include those countries
freq_top20 = freq_df[freq_df['Country'].isin(top_countries)].copy()

# Recalculate codes after filtering for cleaner plotting
freq_top20['CountryCode'] = freq_top20['Country'].astype('category').cat.codes
freq_top20['LangCode'] = freq_top20['LanguageAdmired'].astype('category').cat.codes

# Normalize bubble sizes
freq_top20['Size'] = freq_top20['Count'] / freq_top20['Count'].max() * 1000

# Plotting
plt.figure(figsize=(20, 15))
scatter = plt.scatter(
    x=freq_top20['CountryCode'],
    y=freq_top20['LangCode'],
    s=freq_top20['Size'],
    alpha=0.6,
    c=freq_top20['Count'],
    cmap='plasma'
)

# Updated axis labels for filtered data
country_labels = freq_top20['Country'].astype('category').cat.categories
lang_labels = freq_top20['LanguageAdmired'].astype('category').cat.categories

plt.xticks(ticks=range(len(country_labels)), labels=country_labels, rotation=90)
plt.yticks(ticks=range(len(lang_labels)), labels=lang_labels)

plt.xlabel('Country (Top 20)')
plt.ylabel('Admired Technology')
plt.title('Admired Technologies Across Top 20 Countries (Bubble Size = Frequency)')

cbar = plt.colorbar(scatter)
cbar.set_label('Number of Respondents')

plt.grid(True)
plt.tight_layout()
plt.show()
```
