### Step 2: Load the Dataset


```python
# Load the dataset from the given URL
file_url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/n01PQ9pSmiRX6520flujwQ/survey-data.csv"
df = pd.read_csv(file_url)

# Display the first few rows to understand the structure of the dataset
df.head()
```

<h3>Step 3: Analyze and Visualize Compensation Distribution</h3>


**Task**: Plot the distribution and histogram for `ConvertedCompYearly` to examine the spread of yearly compensation among respondents.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Drop NaN values for plotting
comp_data = df['ConvertedCompYearly'].dropna()

plt.figure(figsize=(14,6))

# Histogram (left)
plt.subplot(1, 2, 1)
sns.histplot(comp_data, bins=10, kde=False, color='skyblue')
plt.title('Histogram of Converted Yearly Compensation')
plt.xlabel('Yearly Compensation')
plt.ylabel('Count')

# Distribution plot with KDE (right)
plt.subplot(1, 2, 2)
sns.kdeplot(comp_data, fill=True, color='purple')
plt.title('Distribution Plot of Converted Yearly Compensation')
plt.xlabel('Yearly Compensation')
plt.ylabel('Density')

plt.tight_layout()
plt.show()
```

<h3>Step 4: Calculate Median Compensation for Full-Time Employees</h3>


**Task**: Filter the data to calculate the median compensation for respondents whose employment status is "Employed, full-time."


```python
## Write your code here
# Filter rows where Employment is 'Employed, full-time' and ConvertedCompYearly is not NaN
full_time_emp = df[(df['Employment'] == 'Employed, full-time') & (df['ConvertedCompYearly'].notna())]

# Calculate median compensation
median_comp = full_time_emp['ConvertedCompYearly'].median()

print(f"Median yearly compensation for full-time employed respondents: ${median_comp:,.2f}")
```

<h3>Step 5: Analyzing Compensation Range and Distribution by Country</h3>


Explore the range of compensation in the ConvertedCompYearly column by analyzing differences across countries. Use box plots to compare the compensation distributions for each country to identify variations and anomalies within each region, providing insights into global compensation trends.



```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Optional: Filter countries with sufficient data (e.g., at least 100 respondents)
country_counts = df['Country'].value_counts()
top_countries = country_counts[country_counts >= 100].index

# Filter dataset for those countries and drop NaNs in compensation and country
df_filtered = df[(df['Country'].isin(top_countries)) & 
                 (df['ConvertedCompYearly'].notna())]

plt.figure(figsize=(15,8))
sns.boxplot(x='Country', y='ConvertedCompYearly', data=df_filtered)

plt.xticks(rotation=90)
plt.title('Distribution of Yearly Compensation by Country')
plt.ylabel('Yearly Compensation (USD)')
plt.xlabel('Country')

plt.yscale('log')  # Optional: Use log scale to handle wide range and outliers

plt.tight_layout()
plt.show()
```

<h3>Step 6: Removing Outliers from the Dataset</h3>


**Task**: Create a new DataFrame by removing outliers from the `ConvertedCompYearly` column to get a refined dataset for correlation analysis.


```python
## Write your code here
# Calculate Q1, Q3, and IQR
Q1 = df['ConvertedCompYearly'].quantile(0.25)
Q3 = df['ConvertedCompYearly'].quantile(0.75)
IQR = Q3 - Q1

# Define bounds for non-outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Filter to keep only rows without outliers in ConvertedCompYearly
df_refined = df[(df['ConvertedCompYearly'] >= lower_bound) & (df['ConvertedCompYearly'] <= upper_bound)]

# Optionally, drop rows where ConvertedCompYearly is NaN if you want a strict dataset
df_refined = df_refined.dropna(subset=['ConvertedCompYearly'])

print(f"Original dataset size: {df.shape[0]}")
print(f"Refined dataset size after outlier removal: {df_refined.shape[0]}")
```

<h3>Step 7: Finding Correlations Between Key Variables</h3>


**Task**: Calculate correlations between `ConvertedCompYearly`, `WorkExp`, and `JobSatPoints_1`. Visualize these correlations with a heatmap.


```python
## Write your code here
import pandas as pd

# Select relevant columns and drop rows with missing values in any of them
cols = ['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']
df_corr = df[cols].dropna()

# Calculate the Pearson correlation matrix
corr_matrix = df_corr.corr(method='pearson')

print(corr_matrix)
```

<h3>Step 8: Scatter Plot for Correlations</h3>


**Task**: Create scatter plots to examine specific correlations between `ConvertedCompYearly` and `WorkExp`, as well as between `ConvertedCompYearly` and `JobSatPoints_1`.


```python
## Write your code here
import matplotlib.pyplot as plt
import seaborn as sns

# Prepare data: drop rows with NaNs in relevant columns
df_scatter = df[['ConvertedCompYearly', 'WorkExp', 'JobSatPoints_1']].dropna()

plt.figure(figsize=(14, 6))

# Scatter plot: ConvertedCompYearly vs WorkExp
plt.subplot(1, 2, 1)
sns.scatterplot(data=df_scatter, x='WorkExp', y='ConvertedCompYearly', alpha=0.5)
plt.title('Converted Compensation vs Work Experience')
plt.xlabel('Work Experience (years)')
plt.ylabel('Yearly Compensation')
plt.yscale('log')  # Optional: log scale to better visualize spread
plt.grid(True)

# Scatter plot: ConvertedCompYearly vs JobSatPoints_1
plt.subplot(1, 2, 2)
sns.scatterplot(data=df_scatter, x='JobSatPoints_1', y='ConvertedCompYearly', alpha=0.5)
plt.title('Converted Compensation vs Job Satisfaction Points')
plt.xlabel('Job Satisfaction Points')
plt.ylabel('Yearly Compensation')
plt.yscale('log')  # Optional: log scale
plt.grid(True)

plt.tight_layout()
plt.show()
```
